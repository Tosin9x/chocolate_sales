This is an analysis of Chocolate sales for Ages chocolate stores for the year 2022-2023

```python
#import the relevant libraries

import os

from dotenv import load_dotenv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset

load_dotenv()

path = os.getenv('DATA-PATH')

df = pd.read_csv(path)
```

```python
# Overview of the dataset

df.info()

df.describe()

df.shape

df.head(10)
```

```python
## checking for duplicated or null values

df.isnull().sum()
```

fix the null coloumns, the empty coulmns will be dropped for the sake of tiume series analysis(with regards date)

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

the dataset indicates a lack of duplicated values

```python
## fixing the data type of the amount column

df.Amount = df.Amount.astype('float64')

df.Amount.sum()
```

## Product performance

```python
## revenue by product

rev_product = df.Amount.groupby(df['Product']).sum()

rev_product.sort_values(ascending=False)
```

```python
## Shipping by product

shipped_product = df.Boxes_Shipped.groupby(df['Product']).sum()

shipped_product.sort_values(ascending=False)
```

The revenue sum shows "70% dark bar" as the best performing product and the "Mixed assortment box" a close second :
